```python
def cat_fare(x):
    if x<10:
        return "Cheap"
    elif x >=10 and x <=20:
        return "MID"
    else:
        return "Expensive"
df['Cat-fare'] = df['Fare'].apply(cat_fare)
```

```python
df
```

```python
def Euro_INR(x):
    return x*80
df['Euro_Inr'] = df['Fare'].apply(Euro_INR)
```

```python
import pandas as pd

# Example with a string
time_delta = pd.to_timedelta('1 days 2 hours 3 minutes')
print(time_delta)

# Example with a numeric value
time_delta = pd.to_timedelta(5, unit='days')
print(time_delta)

# Example with a series
time_series = pd.Series(['1 days', '2 days', '3 days'])
time_delta_series = pd.to_timedelta(time_series)
print(time_delta_series)
```
