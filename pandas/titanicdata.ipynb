```python
import pandas as pd

def reindex_dataframe(df):
    """
    Re-indexes the DataFrame with a new index starting from 1 and incrementing by 2 for each row.

    Args:
        df (pd.DataFrame): The input DataFrame with columns 'A', 'B', and 'C'.

    Returns:
        pd.DataFrame: The re-indexed DataFrame.
    """
    new_index = pd.RangeIndex(start=1, stop=len(df) * 2, step=2)
    return df.set_index(new_index)

# Example usage:
# Assuming you have a DataFrame 'my_df' with columns 'A', 'B', and 'C':
# new_df = reindex_dataframe(my_df)
```
